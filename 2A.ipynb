```python
import matplotlib.pyplot as plt

# Interpolation Search
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and key >= arr[low] and key <= arr[high]:
        probes += 1

        if arr[low] == arr[high]:
            if arr[low] == key:
                return low, probes
            return -1, probes

        pos = low + ((key - arr[low]) * (high - low)) // (arr[high] - arr[low])

        if arr[pos] == key:
            return pos, probes
        elif arr[pos] < key:
            low = pos + 1
        else:
            high = pos - 1

    return -1, probes


# Binary Search
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1
        mid = (low + high) // 2

        if arr[mid] == key:
            return mid, probes
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1, probes


# Main Program
n = int(input("Enter number of roll numbers: "))

print("Enter sorted roll numbers:")
arr = list(map(int, input().split()))

key = int(input("Enter roll number to search: "))

# Perform Searches
idx_i, probes_i = interpolation_search(arr, key)
idx_b, probes_b = binary_search(arr, key)

# Results
print("\n===== SEARCH RESULTS =====")

if idx_i != -1:
    print(f"Interpolation Search: Found at index {idx_i}")
else:
    print("Interpolation Search: Not Found")

print(f"Interpolation Search Probes = {probes_i}")

if idx_b != -1:
    print(f"Binary Search: Found at index {idx_b}")
else:
    print("Binary Search: Not Found")

print(f"Binary Search Probes = {probes_b}")

# Comparison Table
print("\n===== COMPARISON TABLE =====")
print("----------------------------------------")
print("Algorithm\t\tProbes")
print("----------------------------------------")
print(f"Interpolation Search\t{probes_i}")
print(f"Binary Search\t\t{probes_b}")
print("----------------------------------------")

# Chart
algorithms = ["Interpolation", "Binary"]
probes = [probes_i, probes_b]

plt.figure(figsize=(6, 4))
plt.bar(algorithms, probes)
plt.title("Probe Count Comparison")
plt.xlabel("Search Algorithm")
plt.ylabel("Number of Probes")

for i, v in enumerate(probes):
    plt.text(i, v + 0.05, str(v), ha='center')

plt.show()
```
