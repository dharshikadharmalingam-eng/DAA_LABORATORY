```python
import random
import matplotlib.pyplot as plt

# Interpolation Search for Floating-Point Numbers
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= key <= arr[high]:
        comparisons += 1

        if arr[high] == arr[low]:
            if arr[low] == key:
                return low, comparisons
            return -1, comparisons

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if pos < low or pos > high:
            break

        comparisons += 1

        if abs(arr[pos] - key) < 1e-6:
            return pos, comparisons

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons


# Dataset Sizes
sizes = [10000, 50000, 100000]
comparison_results = []

print("Dataset Size\tComparisons")
print("-" * 30)

for size in sizes:

    # Generate uniformly distributed floating-point numbers
    arr = sorted(random.uniform(0.0, 1000.0) for _ in range(size))

    # Search for an existing element
    key = arr[size // 2]

    index, comparisons = interpolation_search(arr, key)

    comparison_results.append(comparisons)

    print(f"{size}\t\t{comparisons}")

# Plot Comparison Chart
plt.figure(figsize=(8, 5))
plt.plot(sizes, comparison_results, marker='o', linewidth=2)

plt.title("Interpolation Search Performance")
plt.xlabel("Dataset Size")
plt.ylabel("Number of Comparisons")
plt.grid(True)

for x, y in zip(sizes, comparison_results):
    plt.text(x, y, str(y))

plt.show()
```
